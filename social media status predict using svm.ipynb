# convert into numeric value

```python
def convert_to_numeric(value):
    if isinstance(value, str):
        if 'M' in value:
            return float(value.replace('M', '')) * 1e6
        elif 'K' in value:
            return float(value.replace('K', '')) * 1e3
        else:
            return float(value)
    else:
        return float(value)
```

```python
data['Subscribers'] = data['Subscribers'].apply(convert_to_numeric)
data['Views avg.'] = data['Views avg.'].apply(convert_to_numeric)
data['Likes avg.'] = data['Likes avg.'].apply(convert_to_numeric)
data['Comments avg.'] = data['Comments avg.'].apply(convert_to_numeric)
data['Shares avg.'] = data['Shares avg.'].apply(convert_to_numeric)
```

```python
data.head()
```

```python
data['Score'] = data['Subscribers'] + data['Views avg.'] + data['Likes avg.'] + data['Comments avg.'] + data['Shares avg.']
Average_score = data['Score'].mean()
data['Status'] = data['Score'].apply(lambda x: 'High' if x >= Average_score else 'Low')
```

```python
data.head()
```

# Preprocessing

```python
from sklearn import preprocessing 
l=preprocessing.LabelEncoder()
status_encoded=l.fit_transform(data['Status'])
data['Status']=status_encoded
```

# Assigning dependent and independent variable

```python
x=data.iloc[:,3:8]
x
```

```python
y=data.Status
```

```python
y
```

# Splitting the Dataset into Training and Testing set

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

# fitting the Model

```python
from sklearn.svm import SVC
classifier=SVC(kernel='linear',random_state=0)
classifier.fit(x_train,y_train)
```

```python
y_pred=classifier.predict(x_test)
y_pred
```

# Accuracy

```python
from sklearn.metrics import accuracy_score, classification_report
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy}')
```

# Predicting a new value

```python
y_pred=classifier.predict([[13500000,7200000,300000,50000,30000]])
y_pred
if y_pred==0:
    print("High")
if y_pred==1:
    print("Low")
```
